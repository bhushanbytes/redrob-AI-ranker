```python
import pandas as pd 
df = pd.read_json("../data/candidates.jsonl", lines=True)
df.head
```

```python
sample = df.head(10000)

titles = sample["profile"].apply(
    lambda p: p.get("current_title") if isinstance(p, dict) else None
)

titles = titles.dropna()
```

```python
top_30_titles = titles.value_counts().head(30)

print(top_30_titles)
```

```python
skills_df = (
    df.head(10000)["skills"]
      .explode()
      .dropna()
)

skill_names = skills_df.apply(lambda x: x.get("name"))

top_30_skills = skill_names.value_counts().head(30)

top_30_skills
```
